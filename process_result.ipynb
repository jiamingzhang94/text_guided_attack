```python
import os
import json
import sys
sys.path.append('/data2/junhong/proj/text_guide_attack')
import matplotlib.pyplot as plt
```

```python
list_dir=os.listdir("/data2/junhong/proj/text_guide_attack/save_result")
```

```python
result=[]
for name in list_dir:
    with open(os.path.join("/data2/junhong/proj/text_guide_attack/save_result",name),"r",encoding="utf-8") as f:
        data=json.load(f)
        data[0]["epoch"]=int(name.split(".")[0].split("_")[-1])
        result.append(data[0])
```

```python
# 定义一个函数来处理保留四位小数，除了epoch
def round_dict_values_except_epoch(d):
    return {k: round(v, 4) if isinstance(v, float) and k != 'epoch' else v for k, v in d.items()}

# 使用列表推导来处理整个列表
rounded_result = [round_dict_values_except_epoch(item) for item in result]
sorted_result = sorted(rounded_result, key=lambda x: x['epoch'])
sorted_result
result = sorted_result
```

```python
# 提取数据
epochs = [item['epoch'] for item in result]
avg_sim_origin = [item['avg_sim_origin:'] for item in result]
avg_sim_images = [item['avg_sim_images:'] for item in result]
avg_sim_texts = [item['avg_sim_texts:'] for item in result]
avg_sim_image_text = [item['avg_sim_image_text:'] for item in result]
avg_sim_text_image_outputs_encode = [item['avg_sim_text_image_outputs_encode:'] for item in result]

# 画图
plt.figure(figsize=(10, 6))

plt.plot(epochs, avg_sim_origin, label='avg_sim_origin', marker='o')
plt.plot(epochs, avg_sim_images, label='avg_sim_images', marker='o')
plt.plot(epochs, avg_sim_texts, label='avg_sim_texts', marker='o')
plt.plot(epochs, avg_sim_image_text, label='avg_sim_image_text', marker='o')
plt.plot(epochs, avg_sim_text_image_outputs_encode, label='avg_sim_text_image_outputs_encode', marker='o')

plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Average Similarity Metrics Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
